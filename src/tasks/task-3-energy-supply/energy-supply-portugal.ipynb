## Data Cleaning

```python
portugal_df.info()
```

```python
# Drop columns with more than 50% missing values
def drop_columns(portugal_df, threshold):
    for col in df.columns:
        if portugal_df[col].isnull().sum() > threshold:
            portugal_df.drop(col, axis=1, inplace=True)
    return portugal_df

portugal_df = drop_columns(portugal_df, 0.5*len(portugal_df))
print(portugal_df.shape)
print("The number of missing values in the dataset is: ", portugal_df.isnull().sum().sum())
```

```python
# Impute missing values
def impute_missing_values(portugal_df):
    # In case of numerical columns, replace missing values with mean
    for col in portugal_df.columns:
        if portugal_df[col].dtype != 'object':
            portugal_df[col] = portugal_df[col].fillna(portugal_df[col].mean())
    # In case of categorical columns, replace missing values with mode
    for col in portugal_df.columns:
        if portugal_df[col].dtype == 'object':
            portugal_df[col] = portugal_df[col].fillna(portugal_df[col].mode()[0])
    return portugal_df

portugal_df = impute_missing_values(portugal_df)
print("The number of missing values in the dataset is: ", portugal_df.isnull().sum().sum())
```

```python
# A function to plot estimated_generation_gwh by primary_fuel
def plot_estimated_generation_gwh_by_primary_fuel(year):
    portugal_df.groupby('primary_fuel')['estimated_generation_gwh_' + year].sum().plot(kind='bar', figsize=(10, 6), color='purple')
    plt.title('Estimated Generation (GWh) by Primary Fuel Type in ' + year, fontsize=16)
    plt.xticks(rotation=45)
    plt.xlabel('Primary Fuel Type', fontsize=14)
    plt.ylabel('Estimated Generation (GWh)', fontsize=14)
    plt.show();

plot_estimated_generation_gwh_by_primary_fuel('2013')
plot_estimated_generation_gwh_by_primary_fuel('2014')
plot_estimated_generation_gwh_by_primary_fuel('2015')
plot_estimated_generation_gwh_by_primary_fuel('2016')
plot_estimated_generation_gwh_by_primary_fuel('2017')
```
